## Database models

Databases can have different "logical" structures. These logical structures define the manner in which the data is stored, organized, and manipulated.

- Flat
   - Something like a CSV.
   - No structures for indexing or recognising relationships between records
- Hierarchical
    - Data is organised into a tree like structure
    - Each child can only have one parent
    - In order to find data, the tree is traversed from the root node
    - Created by IBM in the 1960s
- Network
    - A flexible way of 
- Relational
    - The now "traditional" way of representing a database.
    - Described using first-order predicate logic
    - Records can have relationships with each other
- Object oriented
    - Information represented in terms of objects
    - Different from relational because relational databases maintain a clearer division between the database model and the application
- Graph
    - Each node is related to other nodes via edges. These edges are stored in the database, making retrieval very fast.
    - Usually NoSQL

## Data Abstraction Levels

There are three data abstraction levels:

1. View level
    - Viewing data through queries, etc.
2. Logical level
    - How the data is stored (see above)
3. Physical level
    - The raw 1s and 0s and hardware configuration

## Database Architecture

### 2 Tier

![images/2-tier.png](images/2-tier.png)

A 2 tier database architecture is based on the "client-server" architecture.

The client directly communicates with the server, which makes it theoretically "quicker" than other architectures which have intermediate applications.

Because the client is effectively querying the database if permissions are not set up right in the DBMS things could go very badly.

### 3 Tier

![images/3-tier.png](images/3-tier.png)

3 tier database architectures are similar to two year, except place a "business logic" layer between the client and data layers.

The client layer is a "dumb" layer. It simply receives data from the business layer and displays it for the end layer.

The business layer contains business logic like validation, calculations, insertion, etc. and acts as an interface between then client and data layers.

The data layer is the database where the datais stored. All database operations eventually go through here.

### N Tier

![images/n-tier.png](images/n-tier.png)

An n-tier database architecture is anything with more than 3 tiers in it.

Some of the things it can contain are:

- Presentation layer
- Application layer
- Business layer
- Data access layer

## Primitive Database Operations

- Union (∪)
- Intersection (∩)
- Diff (-)
- Selection (σ)
    - in `σP(R)`, P is the "selection" predicate
    - P can refer to attributes in R, as well as literal values
    - or comparison operators (>=, <=, ==, !=, >, <, etc,)
    - and can combine multiple predicates (via &&, ||, !=, etc.)
    - Result contains all tuples in R for which P is true
- Projection (Π)
    - in Πa,b...(R): a, b... are the specified attributes of R
    - these attributes must actually be in R
    - results schema only contains these specified attributes
    - result may have fewer rows than R (not all attributes may have values)
- Join (⋈)

## Normalisation

Normalisation is the process of structuring relational databases in such a way that it removes data redundancy and improves data integrity.

### First Normal Form (1NF)

- Each database cell must only contain 1 value
- Each record must be unique

### Second Normal Form (2NF)

- Must be in 1NF
- There must be no partial dependency

### Third Normal Form (3NF)

- Must be in 2NF
- There must be no transitive dependency
