## Database models

Databases can have different "logical" structures. These logical structures define the manner in which the data is stored, organized, and manipulated.

- Flat
   - Something like a CSV.
   - No structures for indexing or recognising relationships between records
- Hierarchical
    - Data is organised into a tree like structure
    - Each child can only have one parent
    - In order to find data, the tree is traversed from the root node
    - Created by IBM in the 1960s
- Network
    - A flexible way of 
- Relational
    - The now "traditional" way of representing a database.
    - Described using first-order predicate logic
    - Records can have relationships with each other
- Object oriented
    - Information represented in terms of objects
    - Different from relational because relational databases maintain a clearer division between the database model and the application
- Graph
    - Each node is related to other nodes via edges. These edges are stored in the database, making retrieval very fast.
    - Usually NoSQL

## Data Abstraction Levels

There are three data abstraction levels:

1. View level
    - Viewing data through queries, etc.
2. Logical level
    - Describes what is stored in the database and what the relationships are
3. Physical level
    - Describes how the data is stored

## Database Architecture

### 2 Tier

![images/2-tier.png](images/2-tier.png)

A 2 tier database architecture is based on the "client-server" architecture.

The client directly communicates with the server, which makes it theoretically "quicker" than other architectures which have intermediate applications.

Because the client is effectively querying the database if permissions are not set up right in the DBMS things could go very badly.

### 3 Tier

![images/3-tier.png](images/3-tier.png)

3 tier database architectures are similar to two year, except place a "business logic" layer between the client and data layers.

The client layer is a "dumb" layer. It simply receives data from the business layer and displays it for the end layer.

The business layer contains business logic like validation, calculations, insertion, etc. and acts as an interface between then client and data layers.

The data layer is the database where the datais stored. All database operations eventually go through here.

### N Tier

![images/n-tier.png](images/n-tier.png)

An n-tier database architecture is anything with more than 3 tiers in it.

Some of the things it can contain are:

- Presentation layer
- Application layer
- Business layer
- Data access layer

## Primitive Database Operations

- Union (∪)
- Intersection (∩)
- Diff (-)
- Selection (σ)
    - in `σP(R)`, P is the "selection" predicate
    - P can refer to attributes in R, as well as literal values
    - or comparison operators (>=, <=, ==, !=, >, <, etc,)
    - and can combine multiple predicates (via &&, ||, !=, etc.)
    - Result contains all tuples in R for which P is true
- Projection (Π)
    - in Πa,b...(R): a, b... are the specified attributes of R
    - these attributes must actually be in R
    - results schema only contains these specified attributes
    - result may have fewer rows than R (not all attributes may have values)
- Join (⋈)

## Database Design

### Logical

- Gathering business requirements, converting those into a "model"
- Describes the needs of the business, not the database (the needs of the business are then used to establish the needs of the database)
- Gather information about:
    - Business processes
    - Business entities (categories of data)
    - Organisational units
- ER diagrams are created using this information

### Physical

- This is essentially taking the business requirements from the logical design step and converting them into the actual design of a database.
- Tables and columns are created based on entities and attributes that were defined during logical modelling
- Constrains are also defined, including primary, foreign, and check constraints (ie. NOT NULL).
- Often highly tied to the database software the database is designed for

## Normalisation

## CAP Theorem

1. Consistency: every read receives the most recent write, or an error
2. Availability: every request receivees a non error response -- without the guarantee it contains the most recent write
3. Partition tolerance: the system continues to operate despite an arbitrary number of packets being dropped between nodes

The cap theorem states that is impossible to satisfy more than two of the 3 guarantees.

## NoSQL

Why NoSQL over relational DBs?

- High speed read/writes
- High data throughput
- High reliability

... are some of the benefits you could potentially get.

Also, NoSQL means that you'll have more flexibility and less iteration cost.

Because NoSQL is higher speeds you don't need specialised (re: expensive) hardware ot make it run well.

NoSQL is a broad category of product, but each product within the category solves a narrow problem. Often just going with a traditional database is a safer choice than randomly picking a trendy NoSQL technology.

### Cassandra

Cassandra is a NoSQL database heavily focussed on availability. The entire database is spread over multiple nodes.

These nodes can be anywhere, from right next door to the other side of the country. This is helpful for disaster mitigation (ie. storm takes out your east coast data center).

Primary keys have two components:

1. Partition keys: determines how the data is split across the available nodes (same partition key == same node)
2. Clustering keys: the row within the node
